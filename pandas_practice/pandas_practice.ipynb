```python
titanic = titanic.rename(columns={'2urvived': 'Survived'})
```

```python
titanic.columns
```

```python
cols = ["Age", "Fare", "Sex", "sibsp", "Parch", "Pclass", "Embarked", "Survived"]
titanic_clean = titanic[cols]

titanic_clean.head()
```

```python
# Passengers who survived
survivors = titanic_clean[titanic_clean["Survived"] == 1]
survivors.head()
```

```python
# Passengers in first class
first_class = titanic_clean[titanic_clean["Pclass"] == 1]
first_class.head()
```

```python
# Survival count by class
titanic_clean.groupby("Pclass")["Survived"].sum()
```

```python
# Average fare by class
titanic_clean.groupby("Pclass")["Fare"].mean()
```

```python
def age_group(age):
    if pd.isna(age):
        return "Unknown"
    elif age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

titanic_clean["AgeGroup"] = titanic_clean["Age"].apply(age_group)

titanic_clean[["Age", "AgeGroup"]].head()
```

```python
def fare_category(row):
    if row.Fare < 20:
        return "Low"
    elif row.Fare < 50:
        return "Medium"
    else:
        return "High"

titanic_clean["FareCategory"] = titanic_clean.apply(fare_category, axis=1)

titanic_clean[["Fare", "FareCategory"]].head()
```

```python
titanic_clean.groupby("AgeGroup")["Survived"].mean()
```
